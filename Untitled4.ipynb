```python
import pandas as pd

df = pd.read_csv("Churn_Modelling.csv")
df.head()
```

```python
df.drop(['Surname'], axis=1, inplace=True)
```

```python
df.columns
```

```python
df['Exited'].mean()*100
```

```python
df.groupby('Geography')['Exited'].mean()*100
```

```python
df['AgeGroup'] = pd.cut(df['Age'],
                       bins=[18,30,45,60,100],
                       labels=['<30','30-45','46-60','60+'])
```

```python
df.groupby('AgeGroup')['Exited'].mean()*100
```

```python
import matplotlib.pyplot as plt

df.groupby('Geography')['Exited'].mean().plot(kind='bar')
plt.title("Churn by Country")
plt.xlabel("Country")
plt.ylabel("Churn Rate")
plt.show()
```

```python
df.groupby('AgeGroup')['Exited'].mean().plot(kind='bar')
plt.title("Churn by Age Group")
plt.show()
```

```python
df.groupby('NumOfProducts')['Exited'].mean().plot(kind='bar')
plt.title("Churn by Products")
plt.show()
```
